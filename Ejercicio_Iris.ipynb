```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Analizar el formato de las variables
variables_continuas = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
variables_nominales = ['Species']
variables_discretas = ['litros_agua']

# Mostrar el formato de las variables
print("Variables Continuas:")
print(variables_continuas)
print("\nVariables Nominales:")
print(variables_nominales)
print("\nVariables Discretas:")
print(variables_discretas)

# Gráficos de histograma para variables continuas
for variable in variables_continuas:
    # Crea una nueva figura con un tamaño específico
    plt.figure(figsize=(15, 8))

    # Crea un histograma con estimación de la densidad del kernel (KDE) para la variable actual
    sns.histplot(data=iris, x=variable, kde=True)

    # Agrega un título al gráfico
    plt.title(f'Histograma de {variable}')

    # Muestra el gráfico
    plt.show()

# Gráfico de barras para variable nominal (Species)
plt.figure(figsize=(15, 8))
# Cuenta las ocurrencias de cada especie y crea un gráfico de barras
sns.countplot(data=iris, x='Species')

# Agrega un título al gráfico
plt.title(f'Gráfico de barras de Species')

# Muestra el gráfico
plt.show()

# Gráfico de barras para variable discreta (litros_agua)
plt.figure(figsize=(15, 8))
# Cuenta las ocurrencias de cada categoría de litros_agua y crea un gráfico de barras
sns.countplot(data=iris, x='litros_agua')

# Agrega un título al gráfico
plt.title(f'Gráfico de barras de litros_agua')

# Muestra el gráfico
plt.show()

# Gráficos de cajas y bigotes para variables continuas
for variable in variables_continuas:
    # Crea una nueva figura con un tamaño específico
    plt.figure(figsize=(15, 8))

    # Crea un gráfico de cajas y bigotes para la variable actual, separado por la especie
    sns.boxplot(data=iris, x=variable, y='Species')

    # Agrega un título al gráfico
    plt.title(f'Caja y bigotes de {variable} por Species')

    # Muestra el gráfico
    plt.show()

# Gráficos de dispersión para variables continuas con respecto a Species
for i in range(len(variables_continuas)):
    for j in range(i + 1, len(variables_continuas)):
        # Crea una nueva figura con un tamaño específico
        plt.figure(figsize=(15, 8))

        # Crea un gráfico de dispersión para la pareja actual de variables continuas,
        # con colores diferenciados por especie
        sns.scatterplot(data=iris, x=variables_continuas[i], y=variables_continuas[j], hue='Species')

        # Agrega un título al gráfico
        plt.title(f'Dispersión entre {variables_continuas[i]} y {variables_continuas[j]} por Species')

        # Muestra el gráfico
        plt.show()

# Gráficos de dispersión para variables continuas con respecto a litros_agua
for i in range(len(variables_continuas)):
    for j in range(i + 1, len(variables_continuas)):
        # Crea una nueva figura con un tamaño específico
        plt.figure(figsize=(15, 8))

        # Crea un gráfico de dispersión para la pareja actual de variables continuas,
        # con colores diferenciados por valores de litros_agua
        sns.scatterplot(data=iris, x=variables_continuas[i], y=variables_continuas[j], hue='litros_agua')

        # Agrega un título al gráfico
        plt.title(f'Dispersión entre {variables_continuas[i]} y {variables_continuas[j]} por litros_agua')

        # Muestra el gráfico
        plt.show()

# Comentarios sobre el uso analítico de los gráficos:
# - Histogramas: Muestran la distribución y concentración de datos para cada variable continua.
# - Gráfico de Barras para Species: Muestra la proporción de cada especie en el conjunto de datos.
# - Gráfico de Barras para litros_agua: Muestra la distribución de categorías de litros_agua.
# - Gráficos de Cajas y Bigotes: Muestran la distribución y los valores atípicos de variables continuas por especie.
# - Gráficos de Dispersión por Species: Permiten observar las relaciones entre variables continuas para diferentes especies.
# - Gráficos de Dispersión por litros_agua: Permiten observar las relaciones entre variables continuas para diferentes categorías de litros_agua.
```

```python
#Para manejar datos perdidos, podemos eliminar las filas con valores nulos usando dropna(),
#o llenar los valores nulos con la media o mediana de la columna usando fillna().
datos_filtrados = iris.fillna(iris.mean())

#Para eliminar los valores extremos, por ejemplo, podemos utilizar el rango intercuartílico:
for especie in datos_filtrados['Species'].unique():
  for j in variables_continuas:
    # Definir Q1 y Q3 para cada variable continua de cada especie
    Q1 = datos_filtrados.loc[datos_filtrados['Species'] == especie, j].quantile(0.25)
    Q3 = datos_filtrados.loc[datos_filtrados['Species'] == especie, j].quantile(0.75)
    IQR = Q3 - Q1

    # Definir los límites para los valores extremos de cada especie
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Filtrar los datos para cada especie en funcion de variables continuas
    datos_filtrados[j] = datos_filtrados[j][(datos_filtrados['Species'] != especie) | ((datos_filtrados['Species'] == especie) & (datos_filtrados[j] >= limite_inferior) & (datos_filtrados[j] <= limite_superior))]

```

```python
# Gráficos de histograma para variables continuas
for variable in variables_continuas:
    plt.figure(figsize=(15, 8))
    sns.histplot(data=datos_filtrados, x=variable, kde=True)
    plt.title(f'Histograma de {variable}')
    plt.show()

# Gráfico de barras para variables nominales
plt.figure(figsize=(15, 8))
sns.countplot(data=datos_filtrados, x='Species')
plt.title(f'Gráfico de barras de Species')
plt.show()

# Gráfico de barras para variable discreta (litros_agua)
plt.figure(figsize=(15, 8))
sns.countplot(data=datos_filtrados, x='litros_agua')
plt.title(f'Gráfico de barras de litros_agua')
plt.show()

# Gráficos de cajas y bigotes para variables continuas
for variable in variables_continuas:
    plt.figure(figsize=(15, 8))
    sns.boxplot(data=datos_filtrados, x=variable, y='Species')
    plt.title(f'Caja y bigotes de {variable} por Species')
    plt.show()

# Gráficos de dispersión para variables continuas con respecto a Species
for i in range(len(variables_continuas)):
    for j in range(i + 1, len(variables_continuas)):
        plt.figure(figsize=(15, 8))
        sns.scatterplot(data=datos_filtrados, x=variables_continuas[i], y=variables_continuas[j], hue='Species')
        plt.title(f'Dispersión entre {variables_continuas[i]} y {variables_continuas[j]} por Species')
        plt.show()

# Gráficos de dispersión para variables continuas con respecto a litros_agua
for i in range(len(variables_continuas)):
    for j in range(i + 1, len(variables_continuas)):
        plt.figure(figsize=(15, 8))
        sns.scatterplot(data=datos_filtrados, x=variables_continuas[i], y=variables_continuas[j], hue='litros_agua')
        plt.title(f'Dispersión entre {variables_continuas[i]} y {variables_continuas[j]} por litros_agua')
        plt.show()
```
